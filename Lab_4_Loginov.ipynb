1) (0.2 балла) Реализуйте функцию, возвращающую максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой. Для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответом является 5. Если нулевых элементов нет, функция должна возвращать None

```python
def max_el_aft_zer(arr):
    if 0 not in arr:
      return None
    mask = np.where(arr==0)[0]
    arr = np.concatenate((arr, np.array([0])))
    return max(arr[mask+1])
```

```python
def near_val(X, v):
  X = X.flatten()
  X_new = abs(X - v)
  return X[np.where(X_new == min(X_new))][0]
```

Убедитесь, что в функции не будет происходить деления на ноль. Протестируйте на случайной матрице (для её генерации можно использовать, например, функцию numpy.random.randint).

```python
def scale(X):
  X_mean = X.mean(axis=0)
  X_std = X.std(axis=0)
  X_std[X_std == 0] = 0.0001
  return (X - X_mean)/X_std
```

```python
def get_data_mat(X):
    X_det = np.linalg.det(X)
    X_trace = np.linalg.trace(X)
    X_min = np.min(X)
    X_max = np.max(X)
    X_fro = np.linalg.norm(X, 'fro')
    X_lambdas = np.linalg.eig(X)[0]
    X_inv = np.linalg.inv(X)
    return X_det, X_trace, X_min, X_max, X_fro, X_lambdas, X_inv
```

Преобразуйте каждый признак FeatureName из указанных в пару новых признаков FeatureName\_Hour, FeatureName\_Minute, разделив каждое из значений на часы и минуты. Например, признак DepTime необходимо заменить на пару признаков DepTime_Hour, DepTime_Minute. При этом, например, значение 155 исходного признака будет преобразовано в значения 1 и 55 признаков DepTime_Hour, DepTime_Minute соответственно.

```python
def hhmm_2_h(hhmm):
  try:
    hhmm = str(int(hhmm))
    return int(hhmm[:-2])
  except Exception as e:
    pass

def hhmm_2_m(hhmm):
  try:
    hhmm = str(int(hhmm))
    return int(hhmm[-2:])
  except Exception as e:
    pass
```

Реализуйте функцию transform_data, которая принимает на вход DataFrame с признаками и выполняет следующие шаги:

Замена пропущенных значений на нули для вещественных признаков и на строки 'nan' для категориальных. Масштабирование вещественных признаков с помощью StandardScaler. One-hot-кодирование категориальных признаков с помощью DictVectorizer или функции pd.get_dummies. Метод должен возвращать преобразованный DataFrame, который должна состоять из масштабированных вещественных признаков и закодированных категориальных (исходные признаки должны быть исключены из выборки).

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction import DictVectorizer
```

```python
def transform_data(data):
    data = data.copy()
    numeric_features = data.select_dtypes(include=[np.number]).columns.tolist()
    categorical_features = data.select_dtypes(include=['object']).columns.tolist()

    data[numeric_features] = data[numeric_features].fillna(0)
    data[categorical_features] = data[categorical_features].fillna('nan')

    scaler = StandardScaler()
    data[numeric_features] = scaler.fit_transform(data[numeric_features])

    vectorizer = DictVectorizer(sparse=False)
    encoded = vectorizer.fit_transform(data.to_dict(orient='records'))

    return pd.DataFrame(encoded, columns=vectorizer.get_feature_names_out(), index=data.index)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

metrics = {
    'RMSE': 'neg_root_mean_squared_error',
    'MAE': 'neg_mean_absolute_error',
    'R2': 'r2'
}

results = {}

for metric_name, scoring in metrics.items():

    grid_search = GridSearchCV(
        Ridge(),
        {'alpha': alpha_values},
        cv=kf,
        scoring=scoring,
        return_train_score=True
    )

    grid_search.fit(X_train1, y_train1)
    results[metric_name] = grid_search

plt.figure(figsize=(15, 5))

for i, (metric_name, grid_search) in enumerate(results.items(), 1):
    plt.subplot(1, 3, i)

    test_scores = grid_search.cv_results_['mean_test_score']
    train_scores = grid_search.cv_results_['mean_train_score']

    if metric_name in ['RMSE', 'MAE']:
        test_scores = -test_scores  # Преобразуем обратно в положительные значения
        train_scores = -train_scores

    plt.semilogx(alpha_values, test_scores, 'b-', label='Test', linewidth=2)
    plt.semilogx(alpha_values, train_scores, 'r--', label='Train', linewidth=2)
    plt.axvline(grid_search.best_params_['alpha'], color='green', linestyle=':',
                label=f'Best alpha: {grid_search.best_params_["alpha"]:.3f}')

    plt.xlabel('Коэффициент регуляризации')
    plt.ylabel(metric_name)
    plt.title(f'Ridge: {metric_name} vs alpha')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nЛучшие коэффициенты регуляризации:")
for metric_name, grid_search in results.items():
    best_alpha = grid_search.best_params_['alpha']
    best_score = grid_search.best_score_
    if metric_name in ['RMSE', 'MAE']:
        best_score = -best_score  # Преобразуем обратно
    print(f"{metric_name}: alpha = {best_alpha:.4f}, score = {best_score:.4f}")
```
